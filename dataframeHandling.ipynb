```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_table('data/charginosFullData.txt', index_col=False, names=['mass', 'charge', 'p', 'pdgId', 'status', 'pt', 'eta', 'phi', 'decayLength', 'ctau', 'beta', 'gamma', 'eventId', 'eventRun', 'eventLumiBlock', 'foundTrack', 'trackCharge', 'trackPt', 'dR', 'trackPhi', 'trackEta', 'validHits', 'missingInnerHits', 'missingTrackHits', 'missingOuterHits', 'validLayers', 'layersNoMeasurements'])

charginos = df[df['mass'] == 700.048]

hits = df.copy()
hits = hits[hits['mass'] != 700.048]
hits.dropna(inplace=True, axis=1)
hits.rename({'mass' :'hits', 'ctau':'sub', 'ctau*boost':'trackerpart', 'validHits':'layer'}, axis = 1, inplace = True)

charginos.head()
```

```python
def createHitsDataFrame(df):
    tmpDf = pd.DataFrame(data=[[0,0,0,0]], columns=['hit', 'subdetector', 'trackerPart', 'layerDiskWheel'], index=['hit1'])
    dfList = []
    tmpSize = 0

    for i in range(100):
        wholeHit = df.iloc[i,:].values
        hitNumber = df.iloc[i,:].values[0]

        if hitNumber == 0 and tmpSize == 0:
            dfList.append(tmpDf)

        elif hitNumber == 0 and tmpSize != 0:
            dfList.append(tmpDf)
            tmpSize = 0
            tmpDf = pd.DataFrame(data=[[0,0,0,0]], columns=['hit', 'subdetector', 'trackerPart', 'layerDiskWheel'], index=['hit1'])
            dfList.append(tmpDf)

        else:
            if hitNumber == 1:
                if tmpSize != 0:
                    dfList.append(tmpDf)
                tmpSize = 1
                tmpDf = pd.DataFrame(data=[[wholeHit[0],wholeHit[1],wholeHit[2],wholeHit[3]]], columns=['hit', 'subdetector', 'trackerPart', 'layerDiskWheel'], index=['hit1'])

            elif hitNumber > tmpSize:
                tmpDf.loc['hit'+str(hitNumber)] = [wholeHit[0], wholeHit[1], wholeHit[2], wholeHit[3]]
                tmpSize += 1

            else:
                print(tmpDf)
                dfList.append(tmpDf)
                tmpDf = pd.DataFrame(data=[[wholeHit[0],wholeHit[1],wholeHit[2],wholeHit[3]]], columns=['hit', 'subdetector', 'trackerPart', 'layerDiskWheel'], index=['hit1'])
                tmpSize = 1

    dfList.append(tmpDf)
    dfDf = pd.DataFrame({'hitsDf': dfList})

    return dfDf

hitsDF = createHitsDataFrame(hits)
```
