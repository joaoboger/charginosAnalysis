```python
def createHitsDataFrame(df):
    tmpDf = pd.DataFrame(data=[[0,0,0,0]], columns=['hit', 'subdetector', 'trackerPart', 'layerDiskWheel'], index=['hit1'])
    dfList = []
    tmpSize = 0

    for i in range(df.shape[0]):
        wholeHit = df.iloc[i,:].values
        hitNumber = df.iloc[i,:].values[0]

        if hitNumber == 0 and tmpSize == 0:
            dfList.append(tmpDf)

        elif hitNumber == 0 and tmpSize != 0:
            dfList.append(tmpDf)
            tmpSize = 0
            tmpDf = pd.DataFrame(data=[[0,0,0,0]], columns=['hit', 'subdetector', 'trackerPart', 'layerDiskWheel'], index=['hit1'])
            dfList.append(tmpDf)

        else:
            if hitNumber == 1:
                if tmpSize != 0:
                    dfList.append(tmpDf)
                tmpSize = 1
                tmpDf = pd.DataFrame(data=[[wholeHit[0],wholeHit[1],wholeHit[2],wholeHit[3]]], columns=['hit', 'subdetector', 'trackerPart', 'layerDiskWheel'], index=['hit1'])

            elif hitNumber > tmpSize:
                tmpDf.loc['hit'+str(hitNumber)] = [wholeHit[0], wholeHit[1], wholeHit[2], wholeHit[3]]
                tmpSize += 1

            else:
                dfList.append(tmpDf)
                tmpDf = pd.DataFrame(data=[[wholeHit[0],wholeHit[1],wholeHit[2],wholeHit[3]]], columns=['hit', 'subdetector', 'trackerPart', 'layerDiskWheel'], index=['hit1'])
                tmpSize = 1

    dfList.append(tmpDf)
    dfDf = pd.DataFrame({'hitsDf': dfList})

    return dfDf
```

```python
hitsDF = createHitsDataFrame(df)
```

```python
hitsDF.to_pickle('hitsDf.pkl')
```
